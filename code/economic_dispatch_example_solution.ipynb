```python
# import library to calculate a matrix inverse
import numpy as np
# import library to calculate a matrix determinant
import scipy.linalg
# import library to calculate a matrix exponential
```

```python
# define a 3*3 matrix
A = np.array([[0.02, 0, -1], [0, 0.06, -1], [-1, -1, 0]])
# print the matrix
print("Matrix A:")
print(A)
```

```python
# defeine a 3*1 vector Y
Y = np.array([[-20], [-15], [-500]])
print("\nVector Y:")
print(Y)
```

```python
# solution of the linear equation A*X = Y
X = np.linalg.solve(A, Y)
print("\nSolution X of the linear equation A*X = Y:")
print(X)
```
